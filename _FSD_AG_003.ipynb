Question 1 : Create a layout with three boxes side by side using Flexbox. The first box
should have twice the width of the other two. Provide the HTML and CSS code
snippets.



Answer:-

```html
<div class="flex-container">
  <div class="box box-1">Box 1</div>
  <div class="box box-2">Box 2</div>
  <div class="box box-3">Box 3</div>
</div>
```

```css
<style>
.flex-container {
  display: flex;
  width: 100%;
  height: 150px; /* For visualization */
  border: 1px solid #ccc;
}

.box {
  padding: 20px;
  text-align: center;
  color: white;
  font-weight: bold;
  box-sizing: border-box; /* Include padding and border in the element's total width and height */
}

.box-1 {
  flex: 2; /* Takes up 2 parts of the available space */
  background-color: #ff6347; /* Tomato */
}

.box-2,
.box-3 {
  flex: 1; /* Takes up 1 part of the available space */
  background-color: #4682b4; /* SteelBlue */
}
</style>
```

Question 2: Using CSS Grid, create a 2x2 grid layout where the first row spans two
columns, and the second row has two equal columns. Provide the HTML and CSS code
snippets.

Answer:-

```html
<div class="grid-container">
  <div class="grid-item item-1">Header (Spans 2 columns)</div>
  <div class="grid-item item-2">Content A</div>
  <div class="grid-item item-3">Content B</div>
</div>
```

```css
<style>
.grid-container {
  display: grid;
  grid-template-columns: 1fr 1fr; /* Two equal-width columns */
  grid-template-rows: auto auto; /* Two rows, height determined by content */
  gap: 10px; /* Space between grid items */
  width: 100%;
  max-width: 600px; /* Max width for better visualization */
  border: 1px solid #ccc;
  padding: 10px;
  box-sizing: border-box;
}

.grid-item {
  background-color: #5f9ea0; /* CadetBlue */
  color: white;
  padding: 20px;
  text-align: center;
  font-weight: bold;
  box-sizing: border-box;
  border-radius: 5px;
}

.item-1 {
  grid-column: 1 / span 2; /* Starts at column 1 and spans 2 columns */
  background-color: #ff6347; /* Tomato */
}

/* item-2 and item-3 will automatically fill the second row with equal width */
.item-2 {
  background-color: #4682b4; /* SteelBlue */
}

.item-3 {
  background-color: #3cb371; /* MediumSeaGreen */
}
</style>
```

Question 3: Write a media query that changes the background color of the body to light
grey when the screen width is less than 600px.


Answer:-

```css
<style>
body {
  background-color: white; /* Default background color */
}

@media screen and (max-width: 599px) {
  body {
    background-color: lightgrey; /* Background color when screen width is less than 600px */
  }
}
</style>
```

Question 4: Create a button that changes its background color to blue and scales up
slightly when hovered over, using CSS transitions. Provide the HTML and CSS code
snippets.

```html
<button class="hover-button">Hover Me!</button>
```

```css
<style>
.hover-button {
  padding: 15px 30px;
  font-size: 18px;
  background-color: #007bff; /* Default blue background */
  color: white;
  border: none;
  border-radius: 5px;
  cursor: pointer;
  transition: background-color 0.3s ease, transform 0.3s ease; /* Smooth transition for background and transform */
}

.hover-button:hover {
  background-color: #0056b3; /* Darker blue on hover */
  transform: scale(1.05); /* Scales up to 105% of its original size */
}
</style>
```

Question 5: Use the ::before pseudo-element to add a decorative icon before each list
item in an unordered list with the class name ‘decorated-list’. Provide the HTML and
CSS code snippets.

Answer:-

```html
<ul class="decorated-list">
  <li>List Item 1</li>
  <li>List Item 2</li>
  <li>List Item 3</li>
</ul>
```

```css
<style>
.decorated-list {
  list-style: none; /* Remove default list bullet */
  padding-left: 0; /* Adjust padding as list-style: none removes default padding */
}

.decorated-list li {
  position: relative;
  padding-left: 25px; /* Make space for the icon */
  margin-bottom: 10px;
}

.decorated-list li::before {
  content: '▶'; /* Unicode character for a right-pointing triangle. You can use any character or even a font-awesome icon unicode */
  position: absolute;
  left: 0;
  color: #28a745; /* Green color for the icon */
  font-size: 1.2em;
  line-height: 1;
}
</style>
```

Question 6: Create a simple animation that moves a square box from left to right
infinitely. Provide the HTML and CSS code snippets.


Answer:-

```html
<div class="animated-box"></div>
```

```css
<style>
.animated-box {
  width: 100px;
  height: 100px;
  background-color: dodgerblue;
  position: relative; /* Needed for animation to work with 'left' property */
  animation: slide-right 3s infinite alternate;
  border-radius: 10px;
}

@keyframes slide-right {
  0% {
    left: 0;
  }
  100% {
    left: calc(100% - 100px); /* Moves to the right edge, accounting for box width */
  }
}
</style>
```

Question 7: Explain the difference between position: absolute; and position: fixed;
in CSS.


Answer:-

   The main difference between position: absolute; and position: fixed; lies in their positioning context and how they behave during scrolling:

* position: absolute;

Positioning Context: An absolutely positioned element is positioned relative to its nearest positioned ancestor (an ancestor with position other than static). If no such ancestor exists, it's positioned relative to the initial containing block (usually the (html) element).
Scrolling Behavior: When you scroll the page, an absolutely positioned element will scroll along with the rest of the document, maintaining its position relative to its positioned ancestor or the document flow.

 * position: fixed;

Positioning Context: A fixed positioned element is positioned relative to the viewport (the browser window itself).
Scrolling Behavior: When you scroll the page, a fixed positioned element stays in the exact same spot on the screen, regardless of scrolling. This is commonly used for navigation bars, footers, or 'scroll-to-top' buttons that need to remain visible at all times.

Question 8: Create a layered layout where a semi-transparent overlay covers an image
using z-index. Provide the HTML and CSS code snippets.

Answer:-


```html
<div class="image-container">
  <img src="https://via.placeholder.com/400x250/87CEEB/FFFFFF?text=Beautiful+Landscape" alt="Landscape" class="background-image">
  <div class="overlay">Semi-transparent Overlay</div>
</div>
```

```css
<style>
.image-container {
  position: relative; /* Establish a positioning context for absolute children */
  width: 400px; /* Or whatever natural image width */
  height: 250px; /* Or whatever natural image height */
  margin: 20px;
  overflow: hidden; /* Ensures overlay doesn't spill out if larger */
}

.background-image {
  width: 100%;
  height: 100%;
  display: block;
  z-index: 1; /* Image is below the overlay */
}

.overlay {
  position: absolute;
  top: 0;
  left: 0;
  width: 100%;
  height: 100%;
  background-color: rgba(0, 0, 0, 0.5); /* Black with 50% opacity */
  color: white;
  display: flex;
  justify-content: center;
  align-items: center;
  font-size: 24px;
  font-weight: bold;
  z-index: 2; /* Overlay is above the image */
}
</style>
```
