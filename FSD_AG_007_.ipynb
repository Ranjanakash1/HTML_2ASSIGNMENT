````python
%%js
// Create a new paragraph element
const newParagraph = document.createElement('p');

// Set the text content of the new paragraph
newParagraph.textContent = 'Welcome to our website!';

// Get the div with the id 'content'
const contentDiv = document.getElementById('content');

// Check if the contentDiv exists before appending
if (contentDiv) {
  // Add the new paragraph to the contentDiv
  contentDiv.appendChild(newParagraph);
  console.log('Paragraph added successfully!');
} else {
  console.error('Div with id "content" not found!');
}
````

### Explanation of `createElement` and `appendChild`

1.  **`document.createElement(tagName)`**:

    *   **Purpose**: This method creates a new HTML element node with the specified tag name. For example, `document.createElement('p')` creates a new `<p>` element, `document.createElement('div')` creates a `<div>`, and so on.
    *   **Return Value**: It returns the newly created element as an object. This element is not yet part of the document's structure (DOM); it exists only in memory.
    *   **Usage in the code**: `const newParagraph = document.createElement('p');` creates an empty paragraph element and assigns it to the `newParagraph` variable.

2.  **`parentNode.appendChild(childNode)`**:

    *   **Purpose**: This method adds a node to the end of the list of children of a specified parent node. It moves the node if it's already in the document, or inserts it if it's new.
    *   **Return Value**: It returns the appended child node.
    *   **Usage in the code**: `contentDiv.appendChild(newParagraph);` takes the `newParagraph` element (which we created and populated with text) and makes it a child of the `contentDiv` element. This effectively inserts the paragraph into the webpage, inside the `div` with the `id="content"`.

```html
<!DOCTYPE html>
```

````python
%%js
// Get the button element by its ID
const myButton = document.getElementById('myButton');

// Check if the button exists
if (myButton) {
  // Add an event listener for when the mouse enters the button area
  myButton.addEventListener('mouseover', function() {
    myButton.style.backgroundColor = 'blue'; // Change background color to blue
    myButton.style.color = 'white';         // Change text color to white
  });

  // Add an event listener for when the mouse leaves the button area
  myButton.addEventListener('mouseout', function() {
    myButton.style.backgroundColor = 'lightgray'; // Revert background color
    myButton.style.color = 'black';             // Revert text color
  });

  console.log('Event listeners added to button!');
} else {
  console.error('Button with id "myButton" not found!');
}
````

### Explanation of the Code:

1.  **`document.getElementById('myButton')`**:
    *   **Purpose**: This method is used to get a reference to an HTML element by its unique `id` attribute. It's a common way to select a specific element from the DOM.
    *   **Usage**: `const myButton = document.getElementById('myButton');` retrieves the button element with the ID "myButton" and stores it in the `myButton` constant.

2.  **`element.addEventListener(event, function)`**:
    *   **Purpose**: This method attaches an event handler to a specified element. When the specified event occurs (e.g., a click, mouseover, keypress), the provided function will be executed.
    *   **`'mouseover'`**: This event fires when the pointer of a pointing device (like a mouse) is moved onto an element.
    *   **`'mouseout'`**: This event fires when the pointer of a pointing device is moved off an element.
    *   **Usage**: We attach two event listeners to `myButton`:
        *   The first one listens for `mouseover` and, when triggered, sets `myButton.style.backgroundColor` to `'blue'` and `myButton.style.color` to `'white'`.
        *   The second one listens for `mouseout` and, when triggered, reverts the `backgroundColor` to `'lightgray'` and `color` to `'black'`.

3.  **`element.style.propertyName`**:
    *   **Purpose**: The `style` property of an HTML element object provides access to the inline style of that element. You can use it to read or modify CSS properties directly via JavaScript.
    *   **Properties**: CSS properties like `background-color`, `color`, `font-size`, `padding`, etc., are accessed as camelCase JavaScript properties (e.g., `backgroundColor` for `background-color`).
    *   **Usage**:
        *   `myButton.style.backgroundColor = 'blue';` directly sets the inline `background-color` style of the button to blue.
        *   `myButton.style.color = 'white';` directly sets the inline `color` (text color) style of the button to white.

This approach directly manipulates the inline styles of the element, which has high specificity. For more complex styling or managing many states, using CSS classes (e.g., `element.classList.add('hover-effect')`) is often preferred, but `element.style` directly fulfills the request's requirement.

Question 3 : Create a button that shows an alert with the message "Button was clicked!" when clicked. Also, add a second event listener to the same button that changes its text to "Clicked!" after being clicked. Explain why addEventListener is better than using onclick.

Answer:-

### HTML Setup (for demonstration purposes)

To make the JavaScript code runnable in a browser environment, you would typically have an HTML structure that includes a button with the specified ID. This is for context, as Colab runs Python.

```html
<!DOCTYPE html>
<html>
<head>
<title>Button Click Event Example</title>
</head>
<body>
  <button id="myClickButton">Click Me!</button>

  <script src="script.js"></script> <!-- Link to your JavaScript file -->
</body>
</html>
```

````python
%%js
// Get the button element by its ID
const myClickButton = document.getElementById('myClickButton');

// Check if the button exists
if (myClickButton) {
  // First event listener: Show an alert
  myClickButton.addEventListener('click', function() {
    alert('Button was clicked!');
  });

  // Second event listener: Change button text
  myClickButton.addEventListener('click', function() {
    myClickButton.textContent = 'Clicked!';
  });

  console.log('Multiple click event listeners added to button!');
} else {
  console.error('Button with id "myClickButton" not found!');
}
````

### Explanation of the Code and `addEventListener` vs. `onclick`

#### Code Explanation:

1.  **`document.getElementById('myClickButton')`**: We first get a reference to the button element using its ID, similar to previous examples.
2.  **`myClickButton.addEventListener('click', function() { ... });`**: We use `addEventListener` to attach two separate functions (event handlers) to the 'click' event of the `myClickButton`.
    *   The first function displays a browser `alert` box with the message "Button was clicked!".
    *   The second function changes the `textContent` property of the button element to "Clicked!".

Both of these functions will execute sequentially when the button is clicked.

#### Why `addEventListener` is Better Than `onclick`:

The primary reason `addEventListener` is preferred over directly assigning to `element.onclick` is its ability to handle **multiple event handlers** for the same event on a single element.

1.  **Multiple Event Handlers**:
    *   **`addEventListener`**: Allows you to attach multiple distinct functions to the same event type (e.g., several functions for a 'click' event) on a single element. All registered event handlers will be executed when the event occurs.
    *   **`onclick` (or `on<event-type>`)**: You can only assign *one* function to `element.onclick`. If you try to assign a second function, it will overwrite the first one, meaning only the last assigned function will execute.

    *Example of `onclick` limitation:*
    ```javascript
    // This will run.
    myButton.onclick = function() { alert('First handler'); };
    // This will overwrite the first one; 'First handler' will NOT run.
    myButton.onclick = function() { alert('Second handler'); };
    ```

2.  **Decoupling of Code**: `addEventListener` promotes better code organization. Different parts of your application can add their own event handlers to an element without needing to know or interfere with other handlers that might already be attached.

3.  **Flexibility and Control**: `addEventListener` offers more control through its optional third argument (an `options` object or a boolean `useCapture`). This allows you to:
    *   Specify whether the event should be handled during the capturing or bubbling phase (`useCapture`).
    *   Define options like `once` (the listener is invoked at most once after being added), `passive` (the listener will never call `preventDefault()`), and `signal` (an `AbortSignal` object to remove the event listener automatically).

4.  **Removal of Event Listeners**: `removeEventListener` allows you to explicitly remove specific event listeners, which is crucial for preventing memory leaks in complex applications, especially with dynamically added or removed elements.

In summary, `addEventListener` is a more robust, flexible, and modern way to handle events in JavaScript, providing better control and preventing unintended overwrites compared to the older `on<event-type>` property assignment.

Question 4 : Write a function named extractDomain that takes an email string like "john@example.com" and returns the domain part ("example.com") using only string methods. Do not use regular expressions.

Answer:-

````python
%%js
function extractDomain(email) {
  // Find the index of the '@' symbol
  const atIndex = email.indexOf('@');

  // If '@' is found, extract the substring after it
  if (atIndex !== -1) {
    return email.substring(atIndex + 1);
  } else {
    // If no '@' is found, the input might not be a valid email format
    return null; // Or throw an error, depending on desired error handling
  }
}

// Test cases
console.log(extractDomain("john@example.com"));   // Expected: "example.com"
console.log(extractDomain("jane.doe@sub.domain.net")); // Expected: "sub.domain.net"
console.log(extractDomain("noatsign.com"));      // Expected: null
console.log(extractDomain("another@test"));     // Expected: "test"
````

### Explanation of the `extractDomain` Function:

1.  **`email.indexOf('@')`**:
    *   **Purpose**: The `indexOf()` string method returns the first index at which a given element can be found in the string, or -1 if it is not present.
    *   **Usage**: We use `email.indexOf('@')` to find the position of the `@` symbol within the `email` string. This symbol acts as the separator between the username and the domain.
    *   **Result**: The result is stored in the `atIndex` variable.

2.  **Conditional Check (`if (atIndex !== -1)`)**:
    *   **Purpose**: We check if `atIndex` is not -1. If it were -1, it would mean that the `@` symbol was not found in the email string, indicating an invalid email format for this function's purpose.

3.  **`email.substring(atIndex + 1)`**:
    *   **Purpose**: The `substring()` string method extracts characters, in a new string, between two specified indices (or from one specified index to the end of the string).
    *   **Usage**: If the `@` symbol is found, we use `email.substring(atIndex + 1)`. The `atIndex + 1` is crucial because we want to start extracting *after* the `@` symbol, not including it. This effectively gives us the domain part of the email.
    *   **Return Value**: The extracted domain string is returned.

4.  **Error Handling (`else { return null; }`)**:
    *   If the `@` symbol is not found, the function returns `null` (or could throw an error) to indicate that a valid domain could not be extracted from the provided string.

Question 5 : Write a function called repeatMessage that takes a message string and a number, and returns the message repeated that many times, separated by a space. Do not use loops. Example: repeatMessage("hello", 3) should return "hello hello hello"

Answer:-

````python
%%js
function repeatMessage(message, count) {
  if (count <= 0) {
    return "";
  }
  // Create an array of 'count' length, fill it with the message, and then join with spaces.
  return Array(count).fill(message).join(' ');
}

// Test cases
console.log(repeatMessage("hello", 3));  // Expected: "hello hello hello"
console.log(repeatMessage("world", 1));  // Expected: "world"
console.log(repeatMessage("test", 0));   // Expected: ""
console.log(repeatMessage("again", 2));  // Expected: "again again"
````

### Explanation of the `repeatMessage` Function:

This function leverages array methods to achieve string repetition without explicit `for` or `while` loops.

1.  **`if (count <= 0) { return ""; }`**:
    *   **Purpose**: This is an edge case check. If the `count` is zero or negative, it means the message should not be repeated, so an empty string is returned.

2.  **`Array(count)`**:
    *   **Purpose**: This creates a new array with a length equal to the `count` parameter. For example, `Array(3)` creates `[empty, empty, empty]`.
    *   **Note**: When an array is created with a specific length using the `Array` constructor (e.g., `new Array(length)`), it creates an array with "empty slots". These empty slots are distinct from `undefined` values and require `fill()` before `map()` or `forEach()` can be used reliably.

3.  **`.fill(message)`**:
    *   **Purpose**: The `fill()` method changes all elements in an array to a static value, from a start index (default 0) to an end index (default array.length).
    *   **Usage**: `Array(count).fill(message)` takes the array of empty slots and fills each slot with the provided `message` string. For `Array(3).fill("hello")`, the result is `["hello", "hello", "hello"]`.

4.  **`.join(' ')`**:
    *   **Purpose**: The `join()` method creates and returns a new string by concatenating all of the elements in an array, separated by commas or a specified separator string.
    *   **Usage**: `.join(' ')` takes the array of repeated messages (e.g., `["hello", "hello", "hello"]`) and concatenates them into a single string, using a space character (`' '`) as the separator between each element. This produces the desired output: `"hello hello hello"`.

### The JavaScript Event Loop Explained

JavaScript is a **single-threaded** language, meaning it can execute only one piece of code at a time. However, it can perform **non-blocking** (asynchronous) operations, thanks to the **Event Loop**. The Event Loop is a crucial part of JavaScript's concurrency model, which allows it to handle tasks like network requests, user interactions, and timers without freezing the main thread.

Here are the key components involved:

1.  **Call Stack (or Execution Stack)**:
    *   This is where synchronous JavaScript code is executed. When a function is called, it's pushed onto the stack. When it returns, it's popped off. JavaScript always executes whatever is on top of the call stack.

2.  **Heap**: This is where memory allocation for variables and objects happens.

3.  **Web APIs (Browser Environment) / Node.js APIs (Node.js Environment)**:
    *   These are functionalities provided by the browser (like `setTimeout`, `DOM API`, `fetch`, `XMLHttpRequest`) or Node.js runtime (like `fs`, `http`). They are *not* part of the JavaScript engine itself but are exposed to JavaScript.
    *   When an asynchronous function (like `setTimeout`) is called, it's pushed onto the call stack. Once executed, it hands over its task to the relevant Web API (e.g., `setTimeout` to the Timer API), and then immediately pops off the call stack, allowing the next synchronous code to run.

4.  **Callback Queue (or Task Queue / Message Queue)**:
    *   When a Web API finishes its assigned task (e.g., a `setTimeout` timer expires, a network request completes), its callback function (the function that was passed to `setTimeout`, `fetch`, etc.) is *not* immediately pushed back onto the call stack. Instead, it's placed in the Callback Queue.

5.  **Event Loop**:
    *   This is the orchestrator. Its sole job is to continuously monitor the Call Stack and the Callback Queue.
    *   If the Call Stack is empty, it takes the first callback function from the Callback Queue and pushes it onto the Call Stack for execution. This process is continuous.

**In essence**: Synchronous code runs first. Asynchronous operations are handed off to Web APIs, and their callbacks wait in the Callback Queue. The Event Loop ensures that these callbacks are only executed *after* the main Call Stack is empty, preventing blocking and allowing the single-threaded JavaScript to appear concurrent.

The output appears in this specific order due to how the JavaScript Event Loop processes synchronous and asynchronous tasks:

1.  **`console.log('Start');`**
    *   This is a synchronous operation. It is immediately pushed onto the Call Stack, executed, and outputs `Start` to the console. The Call Stack is then empty.

2.  **`setTimeout(function() { console.log('Timeout 1'); }, 0);`**
    *   `setTimeout` is an asynchronous function. It's pushed onto the Call Stack, hands its callback (`function() { console.log('Timeout 1'); }`) over to the Web APIs (specifically, the Timer API) with a minimum delay of 0ms, and then immediately pops off the Call Stack. The callback function is now waiting for the timer to expire.
    *   *Important*: Even with a 0ms delay, `setTimeout` callbacks are *never* executed immediately. They are always placed in the Callback Queue and must wait for the Call Stack to be completely empty before the Event Loop picks them up.

3.  **`console.log('Middle');`**
    *   This is another synchronous operation. It's pushed onto the Call Stack, executes, and outputs `Middle`. The Call Stack becomes empty.

4.  **`setTimeout(function() { console.log('Timeout 2'); }, 0);`**
    *   Similar to the first `setTimeout`, this asynchronous call hands its callback (`function() { console.log('Timeout 2'); }`) to the Web APIs with a 0ms delay. This callback is then placed in the Callback Queue, *after* the `Timeout 1` callback, as it was handed off later.

5.  **`console.log('End');`**
    *   This is the final synchronous operation. It's pushed onto the Call Stack, executes, and outputs `End`. The Call Stack is now empty.

6.  **Event Loop in Action**:
    *   At this point, the entire synchronous code has finished executing, and the Call Stack is empty. The Callback Queue contains the two `setTimeout` callbacks, in the order they were triggered: `[Timeout 1 callback, Timeout 2 callback]`.
    *   The Event Loop sees that the Call Stack is empty. It takes the first callback from the Callback Queue (the one for `Timeout 1`) and pushes it onto the Call Stack.

7.  **`console.log('Timeout 1');`**
    *   The `Timeout 1` callback executes, outputs `Timeout 1`, and then pops off the Call Stack. The Call Stack is empty again.

8.  **Event Loop Continues**:
    *   The Event Loop sees the Call Stack is empty and picks up the next callback from the Callback Queue (the one for `Timeout 2`) and pushes it onto the Call Stack.

9.  **`console.log('Timeout 2');`**
    *   The `Timeout 2` callback executes, outputs `Timeout 2`, and then pops off the Call Stack. The Call Stack is empty, and the Callback Queue is also empty, so the program finishes.

This sequence demonstrates that all synchronous code runs to completion before any asynchronous callbacks from the Callback Queue are processed, even if their delay is set to 0ms.

Question 7 : Write a JavaScript function named startCountdown that logs “Time left: X” every second, starting from 5 down to 1, and finally logs “Time’s up!”. Use setInterval and clearInterval only.  

````python
%%js
function startCountdown() {
  let count = 5;
  console.log('Countdown started!');

  // Use setInterval to execute a function every 1000 milliseconds (1 second)
  const countdownInterval = setInterval(() => {
    if (count > 0) {
      console.log(`Time left: ${count}`);
      count--;
    } else {
      console.log("Time's up!");
      // Stop the interval once the countdown is finished
      clearInterval(countdownInterval);
    }
  }, 1000); // 1000 milliseconds = 1 second
}

// Call the function to start the countdown
startCountdown();
````

### Explanation of the `startCountdown` Function and Methods:

1.  **`setInterval(function, delay)`**:
    *   **Purpose**: This method repeatedly calls a function or executes a code snippet, with a fixed time delay between each call. It continues calling the function until `clearInterval` is called or the window is closed.
    *   **`function`**: The function to be executed repeatedly.
    *   **`delay`**: The time, in milliseconds (thousandths of a second), the timer should wait before each execution of the function. In our case, `1000` means 1 second.
    *   **Return Value**: `setInterval` returns a unique interval ID, which can be used later to stop the execution with `clearInterval`.
    *   **Usage in the code**: `const countdownInterval = setInterval(() => { ... }, 1000);` starts the repeated execution of the arrow function every second, and stores its ID in `countdownInterval`.

2.  **`clearInterval(intervalID)`**:
    *   **Purpose**: This method clears a timer set with the `setInterval()` method. It stops the repeated execution of the function.
    *   **`intervalID`**: The ID value returned by `setInterval()` when the timer was created.
    *   **Usage in the code**: `clearInterval(countdownInterval);` is called inside the `setInterval` callback when `count` reaches 0. This stops the timer and prevents the "Time left" messages from being logged indefinitely.

#### How the Countdown Works:

*   An initial `count` variable is set to `5`.
*   `setInterval` is called to set up a repetitive task that runs every 1000ms.
*   Inside the `setInterval`'s callback function:
    *   It first checks if `count` is greater than 0.
    *   If `count` is positive, it logs `Time left: X` and then decrements `count`.
    *   If `count` is 0 (or less, though it won't go below 0 in this logic), it logs `Time's up!` and then calls `clearInterval(countdownInterval)` to stop the `setInterval` from running any further.

### What are Promises in JavaScript?

**Promises** in JavaScript are objects that represent the eventual completion (or failure) of an asynchronous operation and its resulting value. They are a fundamental part of modern JavaScript for handling asynchronous code more cleanly and managing complex asynchronous flows, replacing older methods like callback hell.

A Promise can be in one of three states:

1.  **`pending`**: Initial state, neither fulfilled nor rejected. The asynchronous operation is still in progress.
2.  **`fulfilled` (or `resolved`)**: Meaning that the operation completed successfully. The Promise will then hold a `value`.
3.  **`rejected`**: Meaning that the operation failed. The Promise will then hold a `reason` (an error).

Once a Promise is `fulfilled` or `rejected`, it is said to be `settled`. A settled Promise cannot change its state again.

#### Key advantages of Promises:

*   **Readability**: They make asynchronous code easier to read and write compared to deeply nested callbacks.
*   **Error Handling**: Provide a more structured way to handle errors in asynchronous operations (`.catch()`).
*   **Chaining**: Allow chaining of asynchronous operations (`.then()`) in a predictable order, making it easier to manage sequential asynchronous tasks.


````python
%%js
/**
 * Simulates a login check using a JavaScript Promise.
 * It resolves with user data after 2 seconds if 'loggedIn' is true (simulated),
 * or rejects with an error message if 'loggedIn' is false.
 * @param {boolean} shouldLoginSuccessfully - A flag to determine if the promise should resolve or reject.
 * @returns {Promise<Object>|Promise<string>} A Promise that resolves with user data or rejects with an error message.
 */
function checkLogin(shouldLoginSuccessfully) {
  return new Promise((resolve, reject) => {
    // Simulate network delay of 2 seconds
    setTimeout(() => {
      if (shouldLoginSuccessfully) {
        // Simulate successful login
        const userData = { userId: 'user123', username: 'john.doe', email: 'john@example.com' };
        resolve(userData); // Resolve the promise with user data
      } else {
        // Simulate failed login
        reject('User not found or invalid credentials.'); // Reject the promise with an error message
      }
    }, 2000); // 2000 milliseconds = 2 seconds
  });
}

// --- How to use the Promise ---

// Scenario 1: Successful login
console.log('Attempting to log in successfully...');
checkLogin(true)
  .then(userData => {
    console.log('Login Successful! User Data:', userData);
  })
  .catch(error => {
    console.error('Login Failed:', error);
  });

// Scenario 2: Failed login
console.log('\nAttempting to log in with failure...');
checkLogin(false)
  .then(userData => {
    console.log('This should not be reached for failed login. User Data:', userData);
  })
  .catch(error => {
    console.error('Login Failed (as expected):', error);
  });
````

### Explanation of the `checkLogin` Promise Code:

1.  **`new Promise((resolve, reject) => { ... })`**:
    *   This is how you create a new Promise. The `Promise` constructor takes a single argument: an `executor` function.
    *   The `executor` function itself takes two arguments: `resolve` and `reject`. These are functions provided by the JavaScript engine.
        *   `resolve(value)`: Call this when the asynchronous operation completes successfully. `value` is the result of the operation.
        *   `reject(reason)`: Call this when the asynchronous operation encounters an error. `reason` is typically an `Error` object or a descriptive string.

2.  **`setTimeout(() => { ... }, 2000);`**:
    *   Inside the `executor`, we use `setTimeout` to simulate an asynchronous operation (like a network request to a server). It introduces a 2-second delay before the `resolve` or `reject` function is called.

3.  **Conditional Logic (`if (shouldLoginSuccessfully) { ... } else { ... }`)**:
    *   This `if/else` block simulates the outcome of the login check.
    *   If `shouldLoginSuccessfully` is `true`, it means the login is successful. We create `userData` and then call `resolve(userData)` to fulfill the Promise with this data.
    *   If `shouldLoginSuccessfully` is `false`, it means the login failed. We then call `reject('User not found or invalid credentials.')` to reject the Promise with an error message.

#### How to Consume (Use) a Promise:

*   **`.then(onFulfilled, onRejected)`**: This method is used to schedule a callback function to be executed when the Promise is *fulfilled*. It optionally takes a second callback for when the Promise is *rejected*.
    *   `onFulfilled`: The function executed when the Promise resolves. It receives the resolved `value` as an argument.
*   **`.catch(onRejected)`**: This method is a shorthand for `.then(null, onRejected)`. It's used to handle errors when the Promise is *rejected*. It receives the rejected `reason` as an argument.

In the example:
*   **`checkLogin(true).then(...)`**: This call simulates a successful login. After 2 seconds, the `resolve` function is called, and the callback inside `.then()` receives the `userData` and logs "Login Successful!". The `.catch()` block is skipped.
*   **`checkLogin(false).catch(...)`**: This call simulates a failed login. After 2 seconds, the `reject` function is called, and the callback inside `.catch()` receives the error message and logs "Login Failed!". The `.then()` block is skipped.

### `async/await` Refactored Function with Error Handling

Now, let's rewrite the `getUserDataAndPosts` function using `async/await`.

````python
%%js
/**
 * Fetches user data and their posts using async/await syntax.
 * Includes proper error handling with try...catch.
 * @param {number} userId - The ID of the user to fetch data and posts for.
 * @returns {Promise<{user: Object, posts: Array}>} A promise that resolves with user data and posts,
 *                                               or rejects if any fetching operation fails.
 */
async function getUserDataAndPostsAsync(userId) {
  try {
    // Await the resolution of fetchUserData
    const userData = await fetchUserData(userId);
    console.log(`Fetched user data for ${userData.name}`);

    // Await the resolution of fetchUserPosts using the obtained userId
    const userPosts = await fetchUserPosts(userData.id);
    console.log(`Fetched posts for ${userData.name}`);

    // Return the combined data
    return { user: userData, posts: userPosts };
  } catch (error) {
    // Catch any errors that occur during either await call
    console.error(`Error in getUserDataAndPostsAsync for user ID ${userId}:`, error.message);
    throw error; // Re-throw the error for the caller to handle
  }
}

// --- Demonstration of Usage ---

// Scenario 1: Successful fetch for User 1
console.log('\n--- Attempting to fetch data for User 1 (Success) ---');
getUserDataAndPostsAsync(1)
  .then(result => {
    console.log('Successfully retrieved data for User 1:', result);
  })
  .catch(error => {
    console.error('Failed to retrieve data for User 1:', error.message);
  });

// Scenario 2: Successful fetch for User 2
console.log('\n--- Attempting to fetch data for User 2 (Success) ---');
getUserDataAndPostsAsync(2)
  .then(result => {
    console.log('Successfully retrieved data for User 2:', result);
  })
  .catch(error => {
    console.error('Failed to retrieve data for User 2:', error.message);
  });

// Scenario 3: Failed fetch for a non-existent user (e.g., User 99)
console.log('\n--- Attempting to fetch data for User 99 (Failure) ---');
getUserDataAndPostsAsync(99)
  .then(result => {
    console.log('This should not be reached for User 99:', result);
  })
  .catch(error => {
    console.error('Failed to retrieve data for User 99 (as expected):', error.message);
  });
````

### Explanation of `async/await`

`async/await` is syntactic sugar built on top of Promises, making asynchronous code look and behave more like synchronous code, which significantly improves readability and maintainability.

#### Key Components:

1.  **`async` Keyword**:
    *   You declare a function as `async` to indicate that it will perform asynchronous operations and will always return a Promise. Even if you return a non-Promise value, JavaScript will implicitly wrap it in a resolved Promise.
    *   Example: `async function getUserDataAndPostsAsync(userId) { ... }`

2.  **`await` Keyword**:
    *   The `await` keyword can only be used inside an `async` function.
    *   It *pauses* the execution of the `async` function until the Promise it's waiting for settles (either resolves or rejects).
    *   If the Promise resolves, `await` returns the resolved value. The function then resumes execution from that point.
    *   If the Promise rejects, `await` *throws* the rejected value (the error). This allows us to use standard `try...catch` blocks for error handling in asynchronous code.
    *   Example: `const userData = await fetchUserData(userId);`

#### How it refactors the Promise-based code:

*   **Sequential Execution Clarity**: In the original Promise code, the nesting of `.then()` calls made the sequential flow a bit harder to follow. With `async/await`, `await fetchUserData(userId)` runs first, and only once `userData` is available (i.e., the Promise resolves), does `await fetchUserPosts(userData.id)` execute. This mirrors how you would read synchronous code.

*   **Simplified Error Handling**: Instead of multiple `.catch()` calls or chaining a single `.catch()` at the end of a long `.then()` chain, `async/await` allows you to wrap all `await` calls in a single `try...catch` block. If any `await`ed Promise rejects, the execution immediately jumps to the `catch` block, just like synchronous error handling. This makes error management much more intuitive.

    *   In the example:
        *   If `fetchUserData(userId)` rejects, the `catch` block is immediately executed, and `fetchUserPosts` is never even called.
        *   If `fetchUserData(userId)` resolves, but `fetchUserPosts(userData.id)` then rejects, the `catch` block is also executed.

Overall, `async/await` provides a cleaner, more readable, and easier-to-debug way to write asynchronous JavaScript code, making it the preferred pattern for modern JavaScript development when dealing with Promises.

### Explanation of the `fetch` API Code:

This code uses the modern `fetch` API along with `async/await` for asynchronous operations, which makes the code cleaner and easier to read than traditional Promise chaining.

1.  **`const apiKey` and `const apiUrl`**: These variables hold your API key and the constructed URL for the OpenWeatherMap API endpoint. The `units=metric` parameter is added to get temperatures in Celsius.

2.  **`async function getWeatherData()`**:
    *   The `async` keyword indicates that this function will perform asynchronous operations. It implicitly returns a Promise.

3.  **`try...catch` Block**:
    *   This block is used for robust error handling. Any errors (like network issues or non-200 HTTP responses) encountered during the `await` calls will be caught in the `catch` block.

4.  **`const response = await fetch(apiUrl);`**:
    *   **`fetch(url)`**: This is the core `fetch` API call. It initiates a network request to the specified `apiUrl`. It returns a Promise that resolves to a `Response` object once the headers are received (even if it's an HTTP error like 404).
    *   **`await`**: The `await` keyword pauses the execution of `getWeatherData` until the `fetch` Promise resolves, giving us the `Response` object.
