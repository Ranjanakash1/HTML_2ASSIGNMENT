Question 1 : Create an object named student with the following properties:

● name: "Rahul"

   Then, perform the following operations:
1. Add a new property grade with the value "A".

Answer:-   

### Answer:

Question 2: Write a function checkEligibility that takes an age as a parameter
and returns:

● "Eligible" if age is 18 or above

● "Not Eligible" if age is below 18

Use an if-else statement.


### Answer:

```javascript
function checkEligibility(age) {
    if (age >= 18) {
        return "Eligible";
    }
    else {
        return "Not Eligible";
    }
}

// Test cases
console.log(`Age 20: ${checkEligibility(20)}`); // Expected: Eligible
console.log(`Age 18: ${checkEligibility(18)}`); // Expected: Eligible
console.log(`Age 16: ${checkEligibility(16)}`); // Expected: Not Eligible
```


Question 3: Write a function getDayName that takes a number (1-7) and returns the
corresponding day of the week. Use a switch statement.



### Answer :

```javascript
function getDayName(dayNumber) {
    let dayName;
    switch (dayNumber) {
        case 1:
            dayName = "Sunday";
            break;
        case 2:
            dayName = "Monday";
            break;
        case 3:
            dayName = "Tuesday";
            break;
        case 4:
            dayName = "Wednesday";
            break;
        case 5:
            dayName = "Thursday";
            break;
        case 6:
            dayName = "Friday";
            break;
        case 7:
            dayName = "Saturday";
            break;
        default:
            dayName = "Invalid Day Number";
    }
    return dayName;
}

// Test cases
console.log(`Day 1: ${getDayName(1)}`); // Expected: Sunday
console.log(`Day 3: ${getDayName(3)}`); // Expected: Tuesday
console.log(`Day 7: ${getDayName(7)}`); // Expected: Saturday
console.log(`Day 0: ${getDayName(0)}`); // Expected: Invalid Day Number
```


Question 5: Given an array numbers = [1, 2, 3, 4, 5], perform the following:
1. Use a for loop to print each number.
2. Use a while loop to print each number.
3. Use a for-in loop to print each index.
4. Use a for-of loop to print each number.

### Answer :

```javascript
let numbers = [1, 2, 3, 4, 5];

console.log("--- Using a for loop ---");
for (let i = 0; i < numbers.length; i++) {
    console.log(numbers[i]);
}

console.log("\n--- Using a while loop ---");
let j = 0;
while (j < numbers.length) {
    console.log(numbers[j]);
    j++;
}

console.log("\n--- Using a for-in loop (prints indices) ---");
for (let index in numbers) {
    console.log(index); // Prints the index (0, 1, 2, 3, 4)
}

console.log("\n--- Using a for-of loop (prints values) ---");
for (let number of numbers) {
    console.log(number); // Prints the value (1, 2, 3, 4, 5)
}
```

Question 6: Explain what functions are in JavaScript and provide an example of a
function that adds two numbers.


### Answer (JavaScript Functions):

**What are Functions in JavaScript?**

In JavaScript, a **function** is a block of code designed to perform a particular task. It's a fundamental building block of the language that allows you to organize your code, make it reusable, and improve its readability. Functions can take inputs (called parameters), perform operations, and return an output value.

Key characteristics of JavaScript functions:

*   **Reusability:** You can define a function once and call it multiple times from different parts of your program, avoiding code duplication.
*   **Modularity:** Functions help break down complex problems into smaller, manageable pieces, making the code easier to understand and maintain.
*   **Parameters & Arguments:** Functions can accept data (arguments) passed into them via parameters, allowing them to operate on different values each time they are called.
*   **Return Values:** Functions can return a value back to the code that called them, representing the result of their task.
*   **First-Class Citizens:** In JavaScript, functions are treated like any other variable. They can be assigned to variables, passed as arguments to other functions, and returned from other functions.

**Example of a Function that Adds Two Numbers:**

```javascript
// Function definition
function addNumbers(num1, num2) {
    // The function takes two parameters: num1 and num2
    let sum = num1 + num2; // It calculates their sum
    return sum;             // And returns the result
}

// Calling the function with different arguments
console.log("The sum of 5 and 3 is: " + addNumbers(5, 3));        // Output: The sum of 5 and 3 is: 8
console.log("The sum of 10 and -2 is: " + addNumbers(10, -2));    // Output: The sum of 10 and -2 is: 8
let result = addNumbers(100, 200);
console.log("The sum of 100 and 200 is: " + result);              // Output: The sum of 100 and 200 is: 300
```

Question 7: Convert the add function from Question 6 into an arrow function.

```javascript
// Original function from Question 6 for reference
/*
function addNumbers(num1, num2) {
    let sum = num1 + num2;
    return sum;
}
*/

// Converted to an arrow function
const addNumbersArrow = (num1, num2) => {
    let sum = num1 + num2;
    return sum;
};

// A more concise arrow function (implicit return for single expression)
const addNumbersConciseArrow = (num1, num2) => num1 + num2;

// Test cases for the arrow functions
console.log("The sum of 5 and 3 (arrow) is: " + addNumbersArrow(5, 3));
console.log("The sum of 10 and -2 (arrow) is: " + addNumbersArrow(10, -2));

console.log("The sum of 7 and 1 (concise arrow) is: " + addNumbersConciseArrow(7, 1));
console.log("The sum of 20 and 5 (concise arrow) is: " + addNumbersConciseArrow(20, 5));
```

greet function should call the callback with the name.

### Answer :

```javascript
// Define the greet function that takes a name and a callback
function greet(name, callback) {
    console.log(`Hello, ${name}!`);
    // Call the callback function, passing the name to it
    callback(name);
}

// Define a callback function
function sayGoodbye(personName) {
    console.log(`Goodbye, ${personName}! It was nice meeting you.`);
}

// Another callback function
const customMessage = (personName) => {
    console.log(`Sending a special message to ${personName}.`);
};

// Test cases:
console.log("--- Using sayGoodbye callback ---");
greet("Alice", sayGoodbye);

console.log("\n--- Using customMessage callback ---");
greet("Bob", customMessage);

console.log("\n--- Using an anonymous callback function ---");
greet("Charlie", function(personName) {
    console.log(`Wishing a great day to ${personName}!`);
});

// Using an arrow function as an anonymous callback
console.log("\n--- Using an arrow function as an anonymous callback ---");
greet("David", (personName) => {
    console.log(`Hope you're doing well, ${personName}.`);
});
```

Question 10: Given the string text = " Hello World! ", perform the following:
1. Trim the whitespace.
2. Convert to uppercase.
3. Check if it includes "World".


### Answer :

```javascript
const text = " Hello World! ";

console.log("Original string: '" + text + "'");

// 1. Trim the whitespace.
const trimmedText = text.trim();
console.log("1. Trimmed string: '" + trimmedText + "'");

// 2. Convert to uppercase.
const upperCaseText = trimmedText.toUpperCase();
console.log("2. Uppercase string: '" + upperCaseText + "'");

// 3. Check if it includes "World".
const includesWorld = trimmedText.includes("World");
console.log("3. Does it include 'World'? " + includesWorld); // Expected: true

const includeshello = trimmedText.includes("hello");
console.log("   Does it include 'hello' (case-sensitive)? " + includeshello); // Expected: false

const includesworldLowercase = trimmedText.toLowerCase().includes("world");
console.log("   Does it include 'world' (case-insensitive)? " + includesworldLowercase); // Expected: true
```

Question 11: Given the array numbers = [1, 2, 3, 4, 5], perform the following:
1. Add 6 to the end.
2. Remove the first element.
3. Find the index of 4.
4. Check if all numbers are greater than 0.


### Answer :

```javascript
let numbers = [1, 2, 3, 4, 5];

console.log("Initial array:", numbers);

// 1. Add 6 to the end.
numbers.push(6);
console.log("1. After adding 6 to the end:", numbers); // Expected: [1, 2, 3, 4, 5, 6]

// 2. Remove the first element.
numbers.shift();
console.log("2. After removing the first element:", numbers); // Expected: [2, 3, 4, 5, 6]

// 3. Find the index of 4.
const indexOfFour = numbers.indexOf(4);
console.log("3. Index of 4:", indexOfFour); // Expected: 2

// 4. Check if all numbers are greater than 0.
const allGreaterThanZero = numbers.every(num => num > 0);
console.log("4. Are all numbers greater than 0?", allGreaterThanZero); // Expected: true

// Example for false case:
let negativeNumbers = [-1, 2, 3];
const allGreaterThanZeroFalse = negativeNumbers.every(num => num > 0);
console.log("   Are all numbers in [-1, 2, 3] greater than 0?", allGreaterThanZeroFalse); // Expected: false
```
