Question 1 : Explain what the forEach method does in JavaScript arrays and write code that uses forEach to display each student's name from an array of student objects to the console.

 // Array of student objects const students = [

  { name: "Rahul", age: 20 },

  { name: "Priya", age: 19 },

   { name: "Amit", age: 21 },

   { name: "Sneha", age: 20 } ];


ANSWER:-  

     

The `forEach()` method in JavaScript is an **array iteration method**. It executes a provided callback function once for each array element. It's a clean and readable way to loop over arrays, especially when you need to perform an action for each element without needing to create a new array (like `map()`) or reduce it to a single value (like `reduce()`).

**Key characteristics of `forEach()`:**

*   **Iteration**: It iterates over the elements of an array from the first to the last element.
*   **Callback function**: It takes a callback function as an argument, which is executed for each element. This callback function can receive up to three arguments:
    1.  `currentValue`: The current element being processed in the array.
    2.  `index` (optional): The index of the current element being processed.
    3.  `array` (optional): The array `forEach()` was called upon.
*   **No return value**: `forEach()` *always* returns `undefined`. It does not return a new array or modify the original array (unless the callback function explicitly modifies it).
*   **Side effects**: It's primarily used for its side effects, such as logging values, performing calculations, or updating external variables.
*   **Doesn't stop**: Unlike `for` loops, you cannot `break` or `continue` a `forEach` loop. It will always iterate through all elements of the array.

````python
%%javascript

// Array of student objects
const students = [
  { name: "Rahul", age: 20 },
  { name: "Priya", age: 19 },
  { name: "Amit", age: 21 },
  { name: "Sneha", age: 20 }
];

console.log("Student Names:");
students.forEach(student => {
  console.log(student.name);
});
````

Question 2 : You have an array of product prices in rupees. Write code using the map method to create a new array that contains the same prices but with 18% GST added to each price.

const originalPrices = [100, 250, 500, 750, 1000];

Answer:-

  

The `map()` method in JavaScript is another **array iteration method** that is used to transform elements of an array. It creates a **new array** populated with the results of calling a provided function on every element in the calling array.

**Key characteristics of `map()`:**

*   **Transformation**: It's specifically designed for transforming each element of an array into a new value.
*   **Returns a new array**: Unlike `forEach()`, `map()` always returns a **new array**. The original array remains unchanged.
*   **Callback function**: Similar to `forEach()`, it takes a callback function as an argument, which is executed for each element. This callback function can receive up to three arguments:
    1.  `currentValue`: The current element being processed in the array.
    2.  `index` (optional): The index of the current element being processed.
    3.  `array` (optional): The array `map()` was called upon.
*   **Non-mutating**: It is a non-mutating method, meaning it does not modify the original array.

It's very useful when you want to apply a function to every item in a list and get a new list back with the results.

````python
%%javascript

const originalPrices = [100, 250, 500, 750, 1000];
const gstRate = 0.18; // 18% GST

// Use the map method to create a new array with GST added to each price
const pricesWithGST = originalPrices.map(price => {
  const priceWithGST = price * (1 + gstRate);
  return priceWithGST; // Return the new calculated price
});

console.log("Original Prices:", originalPrices);
console.log("Prices with 18% GST:", pricesWithGST);
````

Question 3 : A school wants to find all students who scored more than 80 marks in their exam. The following code has an error that prevents it from working correctly. Fix the code and explain what was wrong.

 const students = [ { name: "Raj", marks: 85 }, { name: "Sita", marks: 92 }, { name: "Vikram", marks: 78 }, { name: "Meera", marks: 88 } ]; const topStudents = students.filter(student => { student.marks > 80; });

Answer:

The `filter()` method in JavaScript is used to **create a new array** containing all elements from the original array that pass a certain test (i.e., for which the provided callback function returns `true`).

**Key characteristics of `filter()`:**

*   **Selection**: It's designed for selecting a subset of elements from an array based on a condition.
*   **Returns a new array**: Like `map()`, `filter()` always returns a **new array**. The original array remains unchanged.
*   **Callback function**: It takes a callback function as an argument, which is executed for each element. This callback function must return a boolean value (`true` or `false`).
    *   If the callback returns `true`, the element is included in the new array.
    *   If the callback returns `false`, the element is excluded.
*   **Non-mutating**: It is a non-mutating method; it does not modify the original array.

It's ideal when you want to get a *filtered* version of an array.

**What was wrong with the original code?**

The original code was:
```javascript
const topStudents = students.filter(student => { student.marks > 80; });
```

The issue here is the **missing `return` statement**. When you use curly braces `{}` in an arrow function, it creates a function block, and you **must explicitly use the `return` keyword** to return a value from that block. If you omit `return`, the function implicitly returns `undefined`.

Since `filter()` expects its callback to return a boolean (`true` or `false`) to determine whether to include an element, and `undefined` is treated as a 'falsy' value in JavaScript, no students were being included in the `topStudents` array.

**The Fix:**

To fix this, you either need to add `return` explicitly:
```javascript
const topStudents = students.filter(student => { return student.marks > 80; });
```
Or, for single-expression arrow functions, you can remove the curly braces and `return` keyword, which implies an automatic return:
```javascript
const topStudents = students.filter(student => student.marks > 80);
```
Both corrected versions will work as intended.

````python
%%javascript

const students = [
  { name: "Raj", marks: 85 },
  { name: "Sita", marks: 92 },
  { name: "Vikram", marks: 78 },
  { name: "Meera", marks: 88 }
];

// Fixed code: added the 'return' keyword to the filter callback
const topStudents = students.filter(student => {
  return student.marks > 80;
});

// Alternative fixed code (more concise for single expressions):
// const topStudents = students.filter(student => student.marks > 80);

console.log("All Students:", students);
console.log("Top Students (marks > 80):");
topStudents.forEach(student => {
  console.log(`- ${student.name} (Marks: ${student.marks})`);
});
````

Question 4 : Write code using the find method to locate a book with the title "Harry Potter" from an array of book objects. If the book is found, display its author name, otherwise display "Book not found".

// Array of book objects const books = [ { title: "Gitanjali", author: "Rabindranath Tagore" }, { title: "Harry Potter", author: "J.K. Rowling" }, { title: "The God of Small Things", author: "Arundhati Roy" }, { title: "Midnight's Children", author: "Salman Rushdie" } ];

Answer:

The `find()` method in JavaScript is used to **return the value of the first element** in the provided array that satisfies the provided testing function. If no elements satisfy the testing function, `undefined` is returned.

**Key characteristics of `find()`:**

*   **Search and Return First Match**: It iterates through the array and returns the *first element* for which the callback function returns a truthy value.
*   **Returns a single element or `undefined`**: Unlike `filter()` which returns an array of matching elements, `find()` returns only the first matching element itself, or `undefined` if no element matches.
*   **Callback function**: It takes a callback function as an argument, which is executed for each element. This callback function must return a boolean-like value.
    *   If the callback returns `true` (or a truthy value), that element is returned immediately, and the iteration stops.
    *   If the callback returns `false` (or a falsy value) for an element, the iteration continues to the next element.
*   **Non-mutating**: It does not modify the original array.

````python
%%javascript

// Array of book objects
const books = [
  { title: "Gitanjali", author: "Rabindranath Tagore" },
  { title: "Harry Potter", author: "J.K. Rowling" },
  { title: "The God of Small Things", author: "Arundhati Roy" },
  { title: "Midnight's Children", author: "Salman Rushdie" }
];

const searchTitle = "Harry Potter";

// Use the find method to locate the book
const foundBook = books.find(book => book.title === searchTitle);

// Display the result
if (foundBook) {
  console.log(`The author of "${searchTitle}" is ${foundBook.author}.`);
} else {
  console.log(`"${searchTitle}" not found.`);
}
````

 Question 5 : Explain how the sort method works in JavaScript and write code to sort an array of employee objects by their salary in descending order (highest salary first).

// Array of employee objects const employees = [ { name: "Rajesh", salary: 45000 }, { name: "Priya", salary: 55000 }, { name: "Amit", salary: 40000 }, { name: "Kavya", salary: 60000 } ];

Answer:-

The `sort()` method in JavaScript is used to **sort the elements of an array in place** and returns the sorted array. The default sort order is ascending, built upon converting the elements into strings, then comparing their sequences of UTF-16 code units.

**Key characteristics of `sort()`:**

*   **In-place modification**: Unlike `map()` and `filter()`, `sort()` **modifies the original array**. It does not create a new sorted array.
*   **Returns the reference**: It returns a reference to the same array that was passed in, now sorted.
*   **Comparison function (optional but highly recommended)**: For sorting anything other than simple strings alphabetically, you should provide a `compareFunction`. This function takes two arguments (`a` and `b`) representing two elements of the array being compared.
    *   If `compareFunction(a, b)` returns a value less than 0, `a` will be sorted before `b`.
    *   If `compareFunction(a, b)` returns a value greater than 0, `b` will be sorted before `a`.
    *   If `compareFunction(a, b)` returns 0, the order of `a` and `b` remains unchanged with respect to each other, but sorted with respect to all other elements.

**Example comparison function for numbers:**
*   **Ascending**: `(a, b) => a - b`
*   **Descending**: `(a, b) => b - a`

````python
%%javascript

// Array of employee objects
const employees = [
  { name: "Rajesh", salary: 45000 },
  { name: "Priya", salary: 55000 },
  { name: "Amit", salary: 40000 },
  { name: "Kavya", salary: 60000 }
];

console.log("Original Employees:", employees);

// Sort the employees array by salary in descending order
employees.sort((a, b) => b.salary - a.salary);

console.log("Employees sorted by salary (descending):");
employees.forEach(employee => {
  console.log(`- ${employee.name}: $${employee.salary}`);
});
````

Question 6 : Write code using the reduce method to calculate the total value of all items in a shopping cart. Each item has a price and quantity. Also find the most expensive single item in the cart.

// Shopping cart with items const cart = [ { name: "Laptop", price: 50000, quantity: 1 }, { name: "Mouse", price: 500, quantity: 2 }, { name: "Keyboard", price: 1500, quantity: 1 }, { name: "Monitor", price: 15000, quantity: 1 } ];

Answer:

The `reduce()` method in JavaScript executes a **reducer** callback function on each element of the array, in order, passing in the return value from the calculation on the preceding element. The final result of running the reducer across all elements of the array is a single value.

**Key characteristics of `reduce()`:**

*   **Aggregation**: It's primarily used for aggregating array elements into a single value (e.g., sum, average, flattening an array, counting occurrences).
*   **Reducer Function**: It takes a reducer function as its first argument. This function takes four arguments:
    1.  `accumulator`: The value resulting from the previous call to the callbackfn. (Or, if `initialValue` is provided, `initialValue`).
    2.  `currentValue`: The value of the current element being processed in the array.
    3.  `currentIndex` (optional): The index of the current element being processed in the array.
    4.  `array` (optional): The array `reduce()` was called upon.
*   **Initial Value (optional)**: It can take an optional `initialValue` as its second argument. This value is used as the first `accumulator` call. If no `initialValue` is provided, the first element of the array is used as the `accumulator`, and `currentValue` starts from the second element.
*   **Returns a single value**: The method returns the single value that results from the reduction.
*   **Non-mutating**: It does not modify the original array (unless the callback explicitly does so, which is generally discouraged).

`reduce()` is incredibly versatile for various data processing tasks.

````python
%%javascript

// Shopping cart with items
const cart = [
  { name: "Laptop", price: 50000, quantity: 1 },
  { name: "Mouse", price: 500, quantity: 2 },
  { name: "Keyboard", price: 1500, quantity: 1 },
  { name: "Monitor", price: 15000, quantity: 1 }
];

// Calculate total value using reduce
const totalCartValue = cart.reduce((accumulator, item) => {
  return accumulator + (item.price * item.quantity);
}, 0); // Initialize accumulator with 0

console.log("Total value of the shopping cart:", totalCartValue + " rupees");

// Find the most expensive single item in the cart using reduce
const mostExpensiveItem = cart.reduce((maxItem, currentItem) => {
  // If maxItem is null (first iteration), or currentItem's price is higher,
  // then currentItem becomes the new maxItem.
  return (maxItem === null || currentItem.price > maxItem.price) ? currentItem : maxItem;
}, null); // Initialize maxItem as null or a dummy object

if (mostExpensiveItem) {
  console.log("Most expensive single item:", mostExpensiveItem.name + " (Price: " + mostExpensiveItem.price + " rupees)");
} else {
  console.log("Cart is empty.");
}
````

Question 7 : Explain what array destructuring is and write code that extracts the first two colors from an array of colors into separate variables using destructuring.

// Array of colors const colors = ["red", "blue", "green", "yellow", "purple"];

Answer:

Array destructuring is a powerful JavaScript feature that allows you to **extract values from arrays** into distinct variables using a syntax that mirrors array literals.

It provides a concise and readable way to 'unpack' values from arrays (or properties from objects) into new variables. This can be particularly useful for:

*   **Assigning multiple variables at once**: Extracting several values from an array into individual variables.
*   **Swapping variable values**: Easily swapping the values of two variables without a temporary variable.
*   **Ignoring unwanted values**: Skipping elements you don't need.
*   **Rest pattern**: Collecting the remaining elements of an array into a new array.

The basic syntax involves using square brackets `[]` on the left-hand side of an assignment to define which variables will receive values from the array on the right-hand side.

````python
%%javascript

// Array of colors
const colors = ["red", "blue", "green", "yellow", "purple"];

console.log("Original colors array:", colors);

// Use array destructuring to extract the first two colors
const [firstColor, secondColor] = colors;

console.log("First color:", firstColor); // Output: red
console.log("Second color:", secondColor); // Output: blue

// You can also skip elements, for example, to get the third color:
const [,, thirdColor] = colors;
console.log("Third color (by skipping):", thirdColor); // Output: green

// Or use the rest pattern to get remaining elements:
const [dominantColor, secondaryColor, ...otherColors] = colors;
console.log("Dominant color:", dominantColor);
console.log("Secondary color:", secondaryColor);
console.log("Other colors:", otherColors); // Output: ["green", "yellow", "purple"]
````

Question 8 : You have a student object with name, age, and marks properties. Write code using object destructuring to extract these values into separate variables and also rename the marks property to score while destructuring.

// Student object const student = { name: "Arjun", age: 20, marks: 85, city: "Mumbai" };

Answer:

Object destructuring is a JavaScript feature that allows you to **extract properties from objects** into distinct variables using a syntax that mirrors object literals. It's a convenient way to unpack values from properties of objects into new variables.

**Key characteristics of object destructuring:**

*   **Extracting properties**: It allows you to bind variables to the same name as an object's properties.
*   **New variables**: The extracted properties are assigned to newly declared variables.
*   **Renaming properties**: You can rename an extracted property to a different variable name during destructuring.
    *   Syntax: `{ originalPropertyName: newVariableName }`
*   **Default values**: You can assign a default value to a variable if the property doesn't exist in the object.
*   **Rest property**: Similar to array destructuring, you can use the rest pattern (`...`) to collect the remaining properties into a new object.
*   **Non-mutating**: It does not modify the original object.

It makes working with object properties much cleaner and more concise.

````python
%%javascript

// Student object
const student = { name: "Arjun", age: 20, marks: 85, city: "Mumbai" };

console.log("Original student object:", student);

// Use object destructuring to extract name, age, and rename marks to score
const { name, age, marks: score } = student;

console.log("Extracted Name:", name);    // Output: Arjun
console.log("Extracted Age:", age);      // Output: 20
console.log("Extracted Score:", score);  // Output: 85 (original marks property renamed to score)

// Trying to access 'marks' directly will result in an error or undefined if not declared
// console.log(marks); // This would cause a ReferenceError if marks was not defined elsewhere

// You can also extract other properties or use the rest operator
const { city, ...otherDetails } = student;
console.log("Extracted City:", city);
console.log("Other details (using rest operator):");
console.log(otherDetails); // Output: { name: "Arjun", age: 20, marks: 85 }
````

Question 9 : Write a function called calculateTotal that can accept any number of prices as arguments and returns their sum. Use the rest operator to handle the variable number of parameters.

Answer:

The **rest operator** (`...`) in JavaScript is a powerful feature introduced in ES6 (ECMAScript 2015) that allows you to collect an **indefinite number of arguments into an array**. It's primarily used in function definitions to handle a variable number of parameters, but it can also be used in array destructuring and object destructuring.

**Key characteristics of the rest operator:**

*   **Collects remaining arguments**: When used in a function's parameter list, it collects all remaining arguments passed to the function into a single array.
*   **Must be the last parameter**: The rest parameter must be the last parameter in a function definition. You cannot have any other parameters after a rest parameter.
*   **Always an array**: Regardless of how many arguments are collected (even zero or one), the rest parameter will always be an array.

**Common use cases:**

*   **Functions with variable arguments**: Creating functions that can accept any number of inputs (like `sum` or `average`).
*   **Destructuring**: Collecting the remaining elements of an array or properties of an object during destructuring assignments.

````python
%%javascript

// Define the calculateTotal function using the rest operator
function calculateTotal(...prices) {
  // 'prices' will be an array containing all arguments passed to the function
  let sum = 0;
  for (let i = 0; i < prices.length; i++) {
    sum += prices[i];
  }
  return sum;

  // Alternatively, using the reduce method:
  // return prices.reduce((accumulator, currentPrice) => accumulator + currentPrice, 0);
}

// Demonstrate the function with different numbers of arguments
console.log("Total for [10, 20, 30]:", calculateTotal(10, 20, 30)); // Output: 60
console.log("Total for [5, 10]:", calculateTotal(5, 10));         // Output: 15
console.log("Total for [100]:", calculateTotal(100));             // Output: 100
console.log("Total for no arguments:", calculateTotal());          // Output: 0
console.log("Total for [15, 25, 35, 45]:", calculateTotal(15, 25, 35, 45)); // Output: 120
````

Question 10 : You have two arrays of student names and need to create a new array that contains all students from both arrays. Also, you have two objects representing a student's basic info and marks. Write code using the spread operator to solve both scenarios.
const classA = ["Rahul", "Priya", "Amit"];  const classB = ["Sneha", "Vikram", "Pooja"];

Answer:

The **spread operator** (`...`) in JavaScript is a feature introduced in ES6 (ECMAScript 2015) that allows an iterable (like an array or string) or an object to be **expanded** in places where zero or more arguments (for function calls) or elements (for array literals) or key-value pairs (for object literals) are expected. It essentially 'spreads out' the elements of an iterable or the properties of an object.

**Key characteristics of the spread operator:**

*   **Expands iterables**: When used with arrays or strings, it unpacks their elements.
*   **Expands objects**: When used with objects, it unpacks their key-value pairs.
*   **Creates shallow copies**: It's often used to create new arrays or objects without mutating the original ones, effectively creating shallow copies.
*   **Concise syntax**: It provides a more readable and concise way to perform operations like concatenating arrays or merging objects compared to older methods.

**Common use cases:**

*   **Combining arrays**: Merging two or more arrays into a new array.
*   **Copying arrays/objects**: Creating a new array or object that is a copy of an existing one.
*   **Adding elements to arrays**: Inserting new elements into an array without `push()` or `unshift()`.
*   **Merging objects**: Combining properties from two or more objects into a new object.
*   **Function calls**: Spreading an array into a function's arguments.

````python
%%javascript

// --- Scenario 1: Combining Arrays ---

const classA = ["Rahul", "Priya", "Amit"];
const classB = ["Sneha", "Vikram", "Pooja"];

console.log("Class A students:", classA);
console.log("Class B students:", classB);

// Use the spread operator to combine both arrays into a new array
const allStudents = [...classA, ...classB];

console.log("All students combined:", allStudents);

console.log("\n---"); // Separator for clarity

// --- Scenario 2: Merging Objects ---

const studentBasicInfo = {
  firstName: "Alice",
  lastName: "Smith",
  age: 21
};

const studentMarks = {
  math: 95,
  science: 88,
  history: 92
};

console.log("Student Basic Info:", studentBasicInfo);
console.log("Student Marks:", studentMarks);

// Use the spread operator to merge basic info and marks into a new student object
const completeStudentProfile = {
  ...studentBasicInfo,
  ...studentMarks,
  id: 'S001' // Can also add new properties
};

console.log("Complete Student Profile:", completeStudentProfile);

// If there are duplicate keys, the last one spread wins:
const obj1 = { a: 1, b: 2 };
const obj2 = { b: 3, c: 4 };
const mergedObj = { ...obj1, ...obj2 };
console.log("Merged object with overlapping keys:", mergedObj); // { a: 1, b: 3, c: 4 }
````
