Question 1 : Explain the purpose of version control systems like Git. Why are they
essential in modern software development?

Answer:-


  Version control systems (VCS) like Git are crucial tools in modern software development. Their primary purpose is to manage changes to source code or any collection of files over time.

Here's a breakdown of their essential roles:

* Tracking Changes: A VCS records every modification made to the codebase, including who made it, when, and why. This creates a complete history of the project.

* Collaboration: Multiple developers can work on the same project simultaneously without overwriting each other's work. VCS allows merging changes from different branches and resolves conflicts.

* Reversion and Rollback: If a bug is introduced or an undesired change is made, developers can easily revert to a previous, stable version of the code.

* Branching and Merging: Developers can create isolated branches to work on new features or bug fixes without affecting the main codebase. Once the work is complete and tested, these branches can be merged back into the main line of development.

 * Accountability and Audit Trail: The history of changes provides an audit trail, showing who made specific modifications, which is valuable for debugging, code reviews, and understanding the evolution of the project.

* Backup and Data Integrity: A VCS acts as a robust backup system, as the entire history of the project is stored, often in distributed repositories, reducing the risk of data loss.

* Deployment and Release Management: VCS helps in managing different versions of software releases, making it easier to deploy specific versions to production and track what changes are included in each release.


Question 2: List the typical steps you would follow when working with Git to make
changes to a project and share them on GitHub.

Answer:-

   Here are the typical steps you would follow when working with Git to make changes to a project and share them on GitHub:

 * Initialize/Clone: Start a new Git repo (git init) or get an existing one (git clone).
* Make Changes: Edit your project files.
* Stage Changes: Add modified files to the staging area (git add .).
* Commit Changes: Save staged changes to history (git commit -m "message").
* Branch (Optional but Recommended): Create a new branch for features/fixes (git branch, git checkout).
* Pull: Get latest changes from the remote repository (git pull origin main).
* Push: Upload your local commits to GitHub (git push origin [branch_name]).
* Pull Request (GitHub): Create a PR on GitHub for code review and merging.




Question 3: What is the purpose of a .gitignore file in a Git repository? Provide an
example of how to ignore all .log files.


Answer:-

The .gitignore file specifies intentionally untracked files that Git should ignore. Files already tracked by Git are not affected by .gitignore. It's commonly used to avoid committing temporary files, build artifacts, operating system specific files, or sensitive information (like API keys) to the repository.

To ignore all .log files, you would add the following line to your .gitignore file:

*.log

Question 7: You are given the following array:
const colors = ["red", "green", "blue", "yellow", "purple"];
1. Access and print the first element of the array.
2. Access and print the last element of the array.
3. Access and print the third element of the array.

Answer:-

```javascript
const colors = ["red", "green", "blue", "yellow", "purple"];

// 1. Access and print the first element:
console.log(colors[0]); // Output: red

// 2. Access and print the last element:
console.log(colors[colors.length - 1]); // Output: purple

// 3. Access and print the third element:
console.log(colors[2]); // Output: blue
```

Question 8: Given x = 10 and y = 5, write expressions to:
1. Add x and y
2. Check if x is greater than y
3. Check if x is equal to y

Answer:-

```javascript
let x = 10;
let y = 5;

// 1. Add x and y
console.log(x + y); // Output: 15

// 2. Check if x is greater than y
console.log(x > y); // Output: true

// 3. Check if x is equal to y
console.log(x === y); // Output: false
```
